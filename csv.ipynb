```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Kullanacağımız csv dosyasını okuyalım


```python
df = pd.read_csv("archive/medicalDoctors.csv")
df
```

Csv dosyasını manipüle ederek istediğimiz bilgilere ulaşalım

```python
european_countries = [
    "Turkey",
    "France",
    "Germany",
    "United Kingdom",
    "Italy",
    "Spain",
    "Portugal",
    "Greece",
    "Netherlands",
    "Belgium",
    "Sweden",
    "Norway",
    "Denmark",
    "Finland",
    "Austria",
    "Switzerland",
    "Poland",
    "Czech Republic",
    "Hungary",
    "Romania",
    "Bulgaria",
    "Croatia",
    "Serbia",
    "Slovakia",
    "Slovenia",
    "Estonia",
    "Latvia",
    "Lithuania",
    "Ukraine",
    "Russia",
    "Belarus",
    "Ireland",
    "Scotland",
    "Wales",
]
european_data = df[df['Location'].isin(european_countries)]
european_data = european_data[european_data["Period"] == 2016]
european_data_sorted = european_data.sort_values(by="First Tooltip", ascending=False)
european_data_sorted
```

Edindiğimiz bilgilerle seaborn kütüphanesini kullanarak bir grafik oluşturalım

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))

sns.barplot(x='First Tooltip', y='Location', data=european_data_sorted, palette='viridis')

plt.xlabel('Medical doctors (per 10,000)')
plt.ylabel('Country')
plt.title('Number of Doctors per 10 Thousand People in European Countries (2016)')
plt.show()
```

```python
df1 = pd.read_csv("archive/WHOregionLifeExpectancyAtBirth.csv")
df1 = df1[df1["Dim1"] == "Both sexes"]
df1
```

```python

plt.xticks(df1["Period"].unique())
locations = df1["Location"].unique()

num_plots = min(len(locations), 3)
plt.figure(figsize=(18, 6))

for i in range(num_plots):
    location = locations[i]
    plt.subplot(1, num_plots, i + 1)  # 1 sıra, num_plots sütunlu alt grafikler oluşturur
    location_data = df1[df1["Location"] == location]
    sns.lineplot(data=location_data, x="Period", y="First Tooltip")
    plt.title(location)
    plt.xlabel("Yıl")
    plt.ylabel("Beklenen Yaşam Süresi (Yıl)")

# Grafikleri göster
plt.tight_layout()
plt.show()
```
