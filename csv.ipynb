Kullanılan kütüphaneler

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Kullanacağımız csv dosyasını okuyalım


```python
df = pd.read_csv("archive/medicalDoctors.csv")
df
```

Csv dosyasını manipüle ederek istediğimiz bilgilere ulaşalım

```python
european_countries = [
    "Turkey",
    "France",
    "Germany",
    "United Kingdom",
    "Italy",
    "Spain",
    "Portugal",
    "Greece",
    "Netherlands",
    "Belgium",
    "Sweden",
    "Norway",
    "Denmark",
    "Finland",
    "Austria",
    "Switzerland",
    "Poland",
    "Czech Republic",
    "Hungary",
    "Romania",
    "Bulgaria",
    "Croatia",
    "Serbia",
    "Slovakia",
    "Slovenia",
    "Estonia",
    "Latvia",
    "Lithuania",
    "Ukraine",
    "Russia",
    "Belarus",
    "Ireland",
    "Scotland",
    "Wales",
]
european_data = df[df['Location'].isin(european_countries)]
european_data = european_data[european_data["Period"] == 2016]
european_data_sorted = european_data.sort_values(by="First Tooltip", ascending=False)
european_data_sorted
```

Edindiğimiz bilgilerle seaborn kütüphanesini kullanarak bir grafik oluşturalım

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 8))

sns.barplot(x='First Tooltip', y='Location', data=european_data_sorted, palette='viridis')

plt.xlabel('Doktor Sayısı  (Her 10 bin)')
plt.ylabel('Ülke')
plt.title('Avrupa ülkelerinde 10 bin kişi başına düşen doktor sayısı (2016)')
plt.show()
```

Başka bir grafik

```python
df1 = pd.read_csv("archive/WHOregionLifeExpectancyAtBirth.csv")
df1 = df1[df1["Dim1"] == "Both sexes"]
df1
```

```python

df1["Period"] = df1["Period"].astype(int)

plt.figure(figsize=(12, 6))
sns.lineplot(data=df1, x="Period", y="First Tooltip", hue="Location")

plt.xticks(df1["Period"].unique())

plt.xlabel("Yıl")
plt.ylabel("Beklenen Yaşam Süresi (Yıl)")
plt.title("Bölgelere Göre Beklenen Yaşam Süresi")


plt.legend(loc='lower right')

plt.show()
```
